```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('Automobile.csv')

data.head()
```

### 1. Tampilkan rata-rata, standar deviasi, nilai minimum, nilai maksimum, Q1, Q2, dan Q3 dari attribute bertipe angka.

```python
numerik_data = data.select_dtypes(include=[np.number])

stats_desc = numerik_data.describe()

Q1 = numerik_data.quantile(0.25)
Q2 = numerik_data.quantile(0.50)  
Q3 = numerik_data.quantile(0.75)

print("Rata-rata:\n", stats_desc.loc['mean'])
print("\nStandar Deviasi:\n", stats_desc.loc['std'])
print("\nNilai Minimum:\n", stats_desc.loc['min'])
print("Nilai Maksimum:\n", stats_desc.loc['max'])
print("Q1 (Kuartil 1):\n", Q1)
print("Q2 (Kuartil 2 / Median):\n", Q2)
print("Q3 (Kuartil 3):\n", Q3)
```

### 2. Tampilkan 10 record pertama dari dataset dengan attribute bertipe angka.

```python
numerical_cols = data.select_dtypes(include=[np.number])
print("\n10 Record Pertama dari Atribut Numerik:\n", numerical_cols.head(10).to_string(index=False))
```

### 3. Tampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka.

```python
num_records = data.shape[0]
num_attributes = data.shape[1]
num_numerical_attributes = len(numerical_cols.columns)
column_names = data.columns
print(f"\nJumlah record: {num_records}")
print(f"Jumlah attribute: {num_attributes}")
print(f"Nama attribute: {list(column_names)}")
print(f"Jumlah attribute numerik: {num_numerical_attributes}")
```

### 4. Tampilkan jumlah label dari semua attribute yang bernilai object

```python
categorical_cols = data.select_dtypes(include=[object])
for col in categorical_cols.columns:
    print(f"\nJumlah label untuk {col}:")
    print(categorical_cols[col].value_counts())
```

### 5. Visualisasikan dataset dengan salah satu grafik. Pilih grafik yang paling cocok untuk memvisualisasikan data tersebut.

```python
plt.figure(figsize=(6, 4))
sns.histplot(data['mpg'], bins=15, kde=True)
plt.title('Distribusi MPG')
plt.xlabel('MPG')
plt.ylabel('Frekuensi')
plt.show()
```

### 6. Tampilkan korelasi dataset menggunakan heatmap.

```python
plt.figure(figsize=(6, 4))
corr = numerical_cols.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap Korelasi')
plt.show()
```
