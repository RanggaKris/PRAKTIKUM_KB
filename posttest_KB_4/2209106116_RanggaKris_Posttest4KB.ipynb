## 1.2 Linear Regression atau Regresi
#### Kita memprediksi mpg berdasarkan horsepower dan weight

```python
X = data[['horsepower', 'weight']]
y = data['mpg']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42) # Split lagi data menjadi train dan test set

# Tangani nilai NaN dengan mengganti NaN dengan rata rata
X_train = X_train.fillna(X_train.mean())
X_test = X_test.fillna(X_test.mean())

lr_model = LinearRegression() # Inisialisasi untuk model regresi linear kita

lr_model.fit(X_train, y_train) # Train model

y_pred = lr_model.predict(X_test) # Predict

print(y_pred) # Hasil predict kita
```

# A2: Evaluasi Hasil Prediksi


## 1. Linear Regression atau Regresi
#### Kita mengevaluasi hasil regresi menggunakan metrik evaluasi yaitu MAE, MSE, RMSE, dan R2 Score

```python
# Evaluasi hasil regresi
mae = mean_absolute_error(y_test, y_pred) # Mean Absolute Error
mse = mean_squared_error(y_test, y_pred) # Mean Squared Error
rmse = np.sqrt(mse) # Root Mean Squared Error
r2 = r2_score(y_test, y_pred) # Coefficient of determination
#R2 fungsinya buat mengukur seberapa baik model menjelaskan variasi dalam data kita

print(f'MAE: {mae}')
print(f'MSE: {mse}')
print(f'RMSE: {rmse}')
print(f'R2 Score: {r2}')
```
